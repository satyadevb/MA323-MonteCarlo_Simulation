```
import numpy as np
import matplotlib.pyplot as plt
import statistics 
import math

a = 1597;b = 20;m = 244944
high = 10000000
u = [];x = 1;i = 0
for i in range(17):
  x = (a*x + b) % m
  u.append(x/m)
for i in range(17,high):
  y = u[i-17] - u[i-5]
  if y < 0:
    y = y + 1
  u.append(y)

freq = []
for i in range(5000):
    freq.append(0)

for i in range(len(u)):
    t = math.ceil(5000 * u[i])
    freq[t-1] = freq[t-1] + 1
for i in range(5000):
    freq[i] = freq[i]/len(u)
print("Probability of variables of 1,3,5,... are ")
num = []
for i in range(5000):
  num.append(2*i - 1)
print(freq)
print("Frequency plot:")
plt.scatter(num,freq)
plt.xlabel('Random number')
plt.ylabel('Frequency probability')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import statistics 
import math
import random

c_val = [135/64,10,20];
high = 1000000
def f(x):
    return (20 * x * (1-x)**3);
    
for c in c_val:
    print("Taking c =",c)
    counts = [];lis = []
    for i in range(high):
        t=0
        while(1):
            x=random.random()
            u=random.random()
            t = t+1
            if u <= f(x)/c:
                lis.append(x)
                counts.append(t)
                break
    print("Average count of iterations = ", statistics.mean(counts))
    x=np.linspace(0,1,200);
    normalised_func = high * f(x)/200;
    print("The plot of Frequency vs Random number:")
    plt.plot(x,normalised_func,color='red');
    plt.hist(lis,bins=200,color = 'b')
    plt.plot()
    plt.xlabel('Random number')
    plt.ylabel('Frequency')
    plt.show()
```
