```
from mpl_toolkits import mplot3d
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import statistics
import random
import math
from scipy import stats
import time

def func_normal(x,y,a):
    return (1/(4*math.pi*math.sqrt((1-a*a))))*math.exp(-0.5*( (x-5)*(x-5) + 4*(y-8)*(y-8) + 4*a*(x-5)*(y-8) ) )

def func1(x):
    return math.exp(-0.5*(x-5)*(x-5)/(1))/(math.sqrt(2*math.pi))

def func2(x):
    return math.exp(-0.5*(x-8)*(x-8)/(4))/(math.sqrt(2*math.pi)*4)


a_list = [-0.5,0,0.5]
    
for a in a_list:
    print("a is",a,"\n")
    X_list = [];Y_list = [];
    for num in range(1000):
      u1 = random.random()
      u2 = random.random()
      r = -2 * math.log(u1)
      v = 2 * math.pi * u2
      z1 = math.sqrt(r) * math.cos(v)
      z2 = math.sqrt(r) * math.sin(v)
      x1 = 5 + z1
      x2 = 8 + 2 * a * z1 + 2 * math.sqrt(1 - a*a) * z2
      X_list.append(x1);Y_list.append(x2)
    print("Frequency plot of simulated density\n")
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    hist, xedges, yedges = np.histogram2d(X_list,Y_list, bins = 50, range=[[0, 10], [3, 13]])
    xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
    xpos = xpos.ravel()
    ypos = ypos.ravel()
    zpos = 0
    dx = dy = 0.5 * np.ones_like(zpos)
    dz = hist.ravel()
    ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')
    plt.show()

    x = np.linspace(0,10,50)
    y = np.linspace(3,13,50)
    X, Y = np.meshgrid(x,y)
    
    func = np.vectorize(func_normal)
    z = func(X, Y, a)

    fig2 = plt.figure()
    ax = plt.axes(projection='3d')
    ax.plot_surface(X, Y, z,linewidth=0)
    ax.set_title("Actual density\n")
    plt.show()
    plt.clf()

    x = np.linspace(0,10, 50)
    y = np.linspace(3, 13, 50)
    freq1 = [0]*50
    freq2 = [0]*50

    for i in X_list:
        if(i>=0 and i <10):
            freq1[int((i)*5)] += 1
    for i in Y_list:
        if(i>=3 and i <13 ):
            freq2[int((i-3)*5)] += 1

    maximum = max(freq1)/func1(5)
    for i in range(50):
        freq1[i]/=maximum
    maximum = max(freq2)/func2(8)
    for i in range(50):
        freq2[i]/=maximum

    m_1 = np.vectorize(func1)
    m_2 = np.vectorize(func2)
    z1 = m_1(x)
    z2 = m_2(y)
    plt.plot(x,z1)
    plt.plot(x, freq1)
    plt.title("Marginal density of X1\n")
    plt.show()

    plt.plot(y, z2)
    plt.plot(y, freq2)
    plt.title("Marginal density of X2\n")
    plt.show()
```
