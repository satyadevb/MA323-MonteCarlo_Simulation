```
import numpy as np
import matplotlib.pyplot as plt
import statistics 
import math
import random
from scipy.special import beta

val = [(1,2),(2,4),(4,2),(3,3),(3,20)]

def f(x,a,b):
    return  ((math.gamma(a+b)/(math.gamma(a)*math.gamma(b))) * x**(a-1) * (1-x)**(b-1))

high = 1000000

for tup in val:
    a1 = tup[0];a2 = tup[1]
    print("\nTaking a1 =",a1,"and a2 =",a2,":")
    x0 = (a1 - 1)/(a1 + a2 - 2)
    print("\n",x0,"is the point where the beta function reaches the maximum")
    c = f(x0,a1,a2)
    u_list = []
    for i in range(high):
      u1 = random.random()
      u2 = random.random()
      if(c*u2 <= f(u1,a1,a2)):
        u_list.append(u1)
    plt.hist(u_list,bins=200,color = 'b')
    plt.plot()
    print("\nThe histogram of the accepted random values :\n")
    plt.xlabel('Random number')
    plt.ylabel('Frequency')
    plt.show()
```
