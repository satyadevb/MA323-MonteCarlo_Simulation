```
import numpy as np
import matplotlib.pyplot as plt
import statistics 
import math
import random
from scipy import stats
import time

def f(x,mean,variance):
    return math.exp((-0.5 * (x-u)**2)/(variance))/(np.sqrt(2*variance*math.pi))

high = [50,5000]

print("Marsaglia-Bray method:")

for num in high:
    start_time = time.time()
    print("\nGenerating",2 * num,"values:\n")
    zlist = [];lis1 = [];lis2 = []
    count = 0
    for i in range(num):
        u1 = random.random()
        u2 = random.random()
        u1 = 2*u1-1
        u2 = 2*u2-1
        count+=1
        x = u1**2 + u2**2
        if(x <= 1):
            y=math.sqrt(-2 * math.log(x)/x)
            z1 = u1*y
            z2 = u2*y
            zlist.append(z1)
            zlist.append(z2)
            lis1.append(math.sqrt(5) * z1)
            lis1.append(math.sqrt(5) * z2)
            lis2.append(math.sqrt(5) * z1 + 5)
            lis2.append(math.sqrt(5) * z2 + 5)
    print("Time taken for execution =",time.time()-start_time)        
    print("Mean is ",statistics.mean(zlist))
    print("Variance is ",statistics.variance(zlist))
    print("\nPlot for N(0,1):")
    histogram, bins = np.histogram(zlist, bins = np.linspace(-5, 5, 50), density=True)
    plt.plot(0.5*(bins[1:] + bins[:-1]), histogram, label="Histogram")
    plt.legend()
    plt.show()
    print("\nPlot for N(0,5):")
    histogram, bins = np.histogram(lis1, bins = np.linspace(-10, 10, 50), density=True)
    distfunc = stats.norm.pdf(0.5*(bins[1:] + bins[:-1]),0,math.sqrt(5))
    plt.plot(0.5*(bins[1:] + bins[:-1]), histogram, label="Histogram")
    plt.plot(0.5*(bins[1:] + bins[:-1]), distfunc,color = 'r',label="Distribution Function")
    plt.legend()
    plt.show() 
    print("\nPlot for N(5,5):")
    histogram, bins = np.histogram(lis2, bins = np.linspace(-5, 15, 50), density=True)
    distfunc = stats.norm.pdf(0.5*(bins[1:] + bins[:-1]),5,math.sqrt(5))
    plt.plot(0.5*(bins[1:] + bins[:-1]), histogram, label="Histogram")
    plt.plot(0.5*(bins[1:] + bins[:-1]), distfunc,color = 'r', label="Distribution Function")
    plt.legend()
    plt.show()
    print("Proportion of Values rejected=",(2*num-len(zlist))/(2*num))
```
