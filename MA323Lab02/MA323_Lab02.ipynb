```
import matplotlib.pyplot as plt

def Fib_gen(a,b,m,num):
  print("Generating",num,"random numbers\n")
  u = []
  x = 1
  for i in range(17):
    x = (a*x + b) % m
    u.append(x/m)
  for i in range(17,num):
    y = u[i-17] - u[i-5]
    if y < 0:
      y = y + 1
    u.append(y)
  v = []
  for i in range(len(u) - 1):
    v.append(u[i+1])
  del u[len(u) - 1]
  print("Plot of (u(i),u(i + 1)) :\n") 
  plt.scatter(u,v)
  plt.show()
  print("Bar diagram is divided into 20 ranges here in [0,1] :")
  plt.hist(u,bins = 20,rwidth = 0.75)
  plt.show()
  print("   ")

m = 244944;a = 1597;b = 20
Fib_gen(a,b,m,1000);
Fib_gen(a,b,m,10000);
Fib_gen(a,b,m,100000)
```

```
import numpy as np
import matplotlib.pyplot as plt
import statistics 
import math

def Exp_random(theta,num):
  u = []
  print("Generating",num,"random numbers :\n")
  for i in range(num):
    k = np.random.uniform(0,1)
    if k!=1:
      y =-theta*(math.log(1-k))
      u.append(y)
  print("Sample mean is :")
  print(statistics.mean(u))
  print("Sample variance is :")
  print(statistics.variance(u))
  print("   ")
  print("Plot of the obtained distribution function : ")
  histo,base = np.histogram(u,bins = 100) 
  cdf = np.cumsum(histo)
  plt.plot(base[:-1],cdf/num,'b')
  plt.xlabel("Random variable value")
  plt.ylabel("Frequency")
  plt.show()
  print("    ")
  print("Plot of the actual distribution function : ")
  x = np.linspace(0,50,100)
  y = 1 - np.exp(-1 * x/theta)
  plt.plot(x,y, 'b')
  plt.xlabel("Random variable value")
  plt.ylabel("Frequency")
  plt.show()
  print("   ")

theta = 5;
Exp_random(theta,1000);
Exp_random(theta,10000);
Exp_random(theta,100000);
```

```
import numpy as np
import matplotlib.pyplot as plt
import statistics 
import math

def Arcsin_random(num):
  u = []
  print("Generating",num,"random numbers :\n")
  for i in range(num):
    k = np.random.uniform(0,1)
    u.append(0.5 * (1 - np.cos(math.pi * k)))
  print("Sample mean is :")
  print(statistics.mean(u))
  print("Sample variance is :")
  print(statistics.variance(u))
  print("   ")
  print("Plot of the obtained distribution function : ")
  histo,base = np.histogram(u,bins = 100) 
  cdf = np.cumsum(histo)
  plt.plot(base[:-1],cdf/num,'b')
  plt.xlabel("Random variable value")
  plt.ylabel("Frequency")
  plt.show()
  print("    ")
  print("Plot of the actual distribution function : ")
  x = np.linspace(0,1,100)
  y = (2 * np.arcsin(np.sqrt(x)))/(math.pi)
  plt.plot(x,y,'b')
  plt.xlabel("Random variable value")
  plt.ylabel("Frequency")
  plt.show()
  print("   ")

Arcsin_random(1000);
Arcsin_random(10000);
Arcsin_random(100000);
```
